### Test and visualise the Image augmentation method

````python
# Function to apply transformations and visualize the result
def visualize_transform(image, transform, num_transformations=3):
    plt.figure(figsize=(10, 5)) # Create Figure
    plt.subplot(1, num_transformations+1, 1)
    plt.imshow(image)
    plt.title('Original Image')
    plt.axis('off') # hide axes

    for i in range(num_transformations):
        # Convert the image to PIL, then apply the transformations
        image_pil = Image.fromarray(image)
        transformed_image = transform(image_pil)
        # Convert torch to numpy for plotting
        transformed_image = transformed_image.permute(1, 2, 0).numpy()
        transformed_image = (transformed_image * np.array([0.229, 0.224, 0.225]) + np.array([0.485, 0.456, 0.406]))
        transformed_image = np.clip(transformed_image, 0, 1) # clip values from 0 to 1.

        plt.subplot(1, num_transformations+1, i+2) # Add subplot for each of the transformations
        plt.imshow(transformed_image)
        plt.title(f'Transformed {i+1}')
        plt.axis('off') # hide axes
    plt.show()
    return

# Apply and display
print('============= Image augment for Train method ================')
for img in images:
    visualize_transform(img, train_transform)
````

## Weight loss calculation
This code helps balance the loss contribution from cancer and non-cancer samples during training, making the model more sensitive to the minority class (cancer).

### Class Weight Calculation Formula

The class weight for each class is calculated as follows:

$$
\text{weight class} = \frac{\text{total samples}}{2 \times \text{num samples in class}}
$$

Where:

*   `total_samples` = the total number of samples in the dataset (sum of all classes)
*   `num_samples_in_class` = the number of samples belonging to the specific class
*   `2` = is a scaling factor that ensures the sum of the weights of all classes, weighted by class frequency, sums up to the number of samples.

<br>

In the context of a binary class example (e.g., cancer/non-cancer), this translates to:

<br>

*   `weight_non_cancer` =  $$ \frac{\text{total samples}}{2 \times \text{num non cancer}} $$
*   `weight_cancer` =  $$ \frac{\text{total samples}}{2 \times \text{num cancer}} $$

reference: https://medium.com/@zergtant/use-weighted-loss-function-to-solve-imbalanced-data-classification-problems-749237f38b75

````python
# Class Weights Calculation
num_cancer = train_data['target'].value_counts()[1]
num_non_cancer = train_data['target'].value_counts()[0]
total_samples = num_cancer + num_non_cancer
weight_non_cancer = total_samples / (2 * num_non_cancer)
weight_cancer = total_samples / (2 * num_cancer)

class_weights = torch.tensor([weight_non_cancer, weight_cancer]).float()
````

## Model CNN (Image only)

### Create MelanomaDataset class for Model CNN (Image only)

1. Loads images and their corresponding labels.
2. Applies data augmentation selectively.
3. Prepares the image data in the correct format so that it can be fed to PyTorch.

````python
class MelanomaDataset(Dataset):
    def __init__(self, dataframe, train_transform=None, val_transform=None):
        self.dataframe = dataframe
        self.train_transform = train_transform
        self.val_transform = val_transform

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        img_path = self.dataframe.iloc[idx]['image_path']
        image = Image.open(img_path).convert('RGB')
        label = self.dataframe.iloc[idx]['target']

        if label == 1:  # Cancerous
           if self.train_transform:
              image = self.train_transform(image)
        else:  # Non-cancerous - we reuse val_transform since we won't augment the validation data
            if self.val_transform:
              image = self.val_transform(image)

        return image, torch.tensor(label, dtype=torch.long)

train_dataset = MelanomaDataset(train_data, train_transform, val_transform)
val_dataset = MelanomaDataset(train_data, train_transform, val_transform)
````

### Define Model CNN (Image only)

````python
# We define a basic CNN model that we can tune later.
# Define Basic CNN
class BasicCNN(nn.Module):
    def __init__(self):
        super(BasicCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 16, kernel_size=3, stride=1, padding=1)
        self.bn1 = nn.BatchNorm2d(16) # Batchnorm added
        self.relu1 = nn.LeakyReLU(0.1) # Leaky ReLU, negative slope 0.1
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2)

        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1)
        self.bn2 = nn.BatchNorm2d(32)
        self.relu2 = nn.LeakyReLU(0.1)
        self.pool2 = nn.MaxPool2d(kernel_size=2, stride=2)

        # Flatten size calculation. Output size = (128/2/2), 32 x 32 for 2 pooling layers
        self.flatten_size = 32 * 32 * 32 # corrected flattening size
        self.fc1 = nn.Linear(self.flatten_size, 64)
        self.relu3 = nn.LeakyReLU(0.1)
        self.dropout = nn.Dropout(0.3) # Dropout added
        self.fc2 = nn.Linear(64, 2)

    def forward(self, x):
        x = self.pool1(self.relu1(self.bn1(self.conv1(x))))
        x = self.pool2(self.relu2(self.bn2(self.conv2(x))))
        x = x.view(x.size(0), -1)
        x = self.dropout(self.relu3(self.fc1(x))) # Dropout
        x = self.fc2(x)
        return x
````

### Training functions for Model CNN (Image only)


**Prepares the model for training.**

1. Loads data in batches from the training set.
2. Performs forward pass to get the model output, and calculates the loss based on the prediction.
3. Backpropagates through the network to update the model parameters.
4. Calculates and returns the average loss and the accuracy for the epoch.

````python
def train_epoch(model, data_loader, optimizer, criterion, device):
    model.train()
    total_loss = 0.0
    all_labels = []
    all_predictions = []

    for images, labels in tqdm(data_loader, desc='Training'): # To see our progress here
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        _, predicted_labels = torch.max(outputs, 1)

        all_labels.extend(labels.cpu().numpy())
        all_predictions.extend(predicted_labels.cpu().numpy())

    avg_loss = total_loss / len(data_loader)
    acc = accuracy_score(all_labels, all_predictions)

    return avg_loss, acc
````

### Validation functions for Model CNN (Image only)

````python
def validate_epoch(model, data_loader, criterion, device, model_name):
    model.eval()
    total_loss = 0.0
    all_labels = []
    all_predictions = []
    all_probs = []

    with torch.no_grad():
        for images, labels in tqdm(data_loader, desc="Validation"):
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            total_loss += loss.item()
            _, predicted_labels = torch.max(outputs, 1)
            all_labels.extend(labels.cpu().numpy())
            all_predictions.extend(predicted_labels.cpu().numpy())
            probs = torch.softmax(outputs, dim=1)[:, 1].cpu().numpy()
            all_probs.extend(probs)


    avg_loss = total_loss / len(data_loader)
    acc = accuracy_score(all_labels, all_predictions)
    auc_roc = roc_auc_score(all_labels, all_probs)
    conf_matrix = confusion_matrix(all_labels, all_predictions)
    precision = precision_score(all_labels, all_predictions, zero_division = 0)
    recall = recall_score(all_labels, all_predictions, zero_division = 0)
    f1 = f1_score(all_labels, all_predictions, zero_division = 0)


    # Print metrics nicely for every epochs
    print(f"\n{model_name} Validation Metrics:")
    print(f"  Validation Loss: {avg_loss:.4f}")
    print(f"  Validation Accuracy: {acc:.4f}")
    print(f"  Validation AUC ROC: {auc_roc:.4f}")
    print(f"  Precision (True Positives): {precision:.4f}") # How many correct predictions of the positive class
    print(f"  Recall (Sensitivity): {recall:.4f}") # How well it predicted the positive class
    print(f"  F1-Score: {f1:.4f}")  # Harmonic mean of precision and recall

    print("  Confusion Matrix:")
    print(f"    True Negative (TN): {conf_matrix[0, 0]}")
    print(f"    False Positive (FP): {conf_matrix[0, 1]}")
    print(f"    False Negative (FN): {conf_matrix[1, 0]}")
    print(f"    True Positive (TP): {conf_matrix[1, 1]}")

    # Plot the confusion matrix for each epoch
    plt.figure(figsize=(6, 5))
    sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues",
                xticklabels=["Non-Cancer", "Cancer"],
                yticklabels=["Non-Cancer", "Cancer"])
    plt.ylabel("True Label")
    plt.xlabel("Predicted Label")
    plt.title(f'{model_name} Confusion Matrix')
    plt.show()


    return avg_loss, acc, auc_roc, precision, recall, f1, conf_matrix
````

### Create MelanomaMetadataDataset class for Model CNN (Image and metadata)

````python
# MelanomaMetadataDataset for Image with Metadata
class MelanomaMetadataDataset(Dataset):
    def __init__(self, dataframe, transform=None, train_transform=None, val_transform=None):
        self.dataframe = dataframe
        self.transform = transform
        self.train_transform = train_transform
        self.val_transform = val_transform
        self.metadata_cols = ['age_encoded', 'sex_encoded', 'anatom_encoded']


    def __len__(self):
        return len(self.dataframe)
    def __getitem__(self, idx):
        img_path = self.dataframe.iloc[idx]['image_path']
        image = Image.open(img_path).convert('RGB')
        label = self.dataframe.iloc[idx]['target']

        if label == 1:  # Cancerous
           if self.train_transform:
              image = self.train_transform(image)
        else:  # Non-cancerous
            if self.val_transform:
              image = self.val_transform(image)

        metadata_values = []
        for col in self.metadata_cols:
           if col == 'age_encoded':
              metadata_values.append(self.dataframe.iloc[idx][col])
           else:
              metadata_values.extend(self.dataframe.iloc[idx][col])

        metadata = np.array(metadata_values,dtype=np.float32) # Convert metadata into numpy first
        metadata = torch.tensor(metadata, dtype=torch.float32) # Then convert metadata into tensor

        return image, metadata, torch.tensor(label, dtype=torch.long)


train_dataset_metadata = MelanomaMetadataDataset(train_data, train_transform=train_transform, val_transform=val_transform)
val_dataset_metadata = MelanomaMetadataDataset(val_data, train_transform=train_transform, val_transform=val_transform)
````

### Define Model CNN (Image and metadata)

````python
# Define CNN with Metadata Model
class CNNImageMetadata(nn.Module):
    def __init__(self, num_metadata_features, num_sex_categories, num_anatom_categories):
        super(CNNImageMetadata, self).__init__()
        # Image processing branch
        self.conv1 = nn.Conv2d(3, 16, kernel_size=3, stride=1, padding=1)
        self.relu1 = nn.ReLU()
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1)
        self.relu2 = nn.ReLU()
        self.pool2 = nn.MaxPool2d(kernel_size=2, stride=2)

        # Calculate the flattened size
        self.flatten_size = 32 * 32 * 32

        # Image feature reduction
        self.fc_image = nn.Sequential(
            nn.Linear(self.flatten_size, 64),
            nn.ReLU()
        )

        # Metadata processing branch
        self.fc_metadata = nn.Sequential(
            nn.Linear(num_metadata_features, 32),
            nn.ReLU(),
            nn.Linear(32, 32),
            nn.ReLU()
        )

        # Combining branches
        self.fc_combine = nn.Sequential(
            nn.Linear(64 + 32, 64),
            nn.ReLU(),
            nn.Linear(64, 2)
        )

    def forward(self, image, metadata):
        # Process image
        x = self.pool1(self.relu1(self.conv1(image)))
        x = self.pool2(self.relu2(self.conv2(x)))
        x = x.view(-1, self.flatten_size)  # Flatten
        image_features = self.fc_image(x)

        # Process metadata
        metadata_features = self.fc_metadata(metadata)

        # Combine features
        combined = torch.cat((image_features, metadata_features), dim=1)
        output = self.fc_combine(combined)

        return output
````
